```python
import torch
from torchvision import datasets, transforms
from torch.utils.data import DataLoader
import torch.nn as nn
```

```python
transform = transforms.Compose([
    transforms.ToTensor()
])
train_dataset = datasets.MNIST(root='./data', train=True, download=True, transform=transform)
test_dataset = datasets.MNIST(root='./data', train=False, download=True, transform=transform)
```

```python
batch_size = 4
train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True)
test_loader = DataLoader(test_dataset, batch_size=batch_size, shuffle=False)
```

```python
conv1 = nn.Conv2d(1, 6, kernel_size=3, stride=1, padding=1)
conv2 = nn.Conv2d(6, 16, kernel_size=3, stride=1, padding=1)
fc1 = nn.Linear(16*7*7, 100)
fc2 = nn.Linear(100, 10)
```

```python
# Collect parameters into a list
parameters = list(conv1.parameters()) + list(conv2.parameters()) + list(fc1.parameters()) + list(fc2.parameters())
```

```python
def forward(self, x):
    x = torch.relu(conv1(x))
    x = torch.max_pool2d(x, 2)
    x = torch.relu(conv2(x))
    x = torch.max_pool2d(x, 2)
    x = x.view(-1, 7 * 7 * 16)
    x = torch.relu(fc1(x))
    x = fc2(x)
    return x
```

```python
criterion = nn.CrossEntropyLoss()
```

```python
for index, (data, label) in enumerate(train_loader):

    x = torch.relu(conv1(data))
    x = torch.max_pool2d(x, 2)
    x = torch.relu(conv2(x))
    x = torch.max_pool2d(x, 2)
    x = x.view(-1, 7 * 7 * 16)
    x = torch.relu(fc1(x))
    x = fc2(x)
    loss = criterion(x, label)
    print(loss.item(), index)
     # backward pass
    for p in parameters:
        p.grad = None
    loss.backward()

    #update
    for p in parameters:
        p.data += -0.01 * p.grad 
    if index == 1000:
        break
```
